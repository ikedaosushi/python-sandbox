```python
import pandas as pd
import numpy as np
import pandas.tseries.offsets as offsets
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('seaborn-dark')
```

```python
cols = [
    'FellAsleepAt','TotalTimeSlept','TotalTimeSleptInSeconds',
    'TimeToFallAsleep','TimeToFallAsleepSeconds','TimeAwake',
    'TimeAwakeSeconds','AwakeCount','RestlessCount','TimeToLeaveBed',
    'TimeToLeaveBedSeconds', 'AwokeAt'
]
```

```python
path = os.path.join(prj_dir, 'data/FitbitSleepLogs.csv')
df = pd.read_csv(path, names=cols, parse_dates=['FellAsleepAt', 'AwokeAt'])
```

```python
df.index = df['FellAsleepAt'].dt.date
df.index.name = 'date'
```
