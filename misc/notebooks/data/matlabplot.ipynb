```python
import matplotlib.pyplot as plt
plt.xkcd()
```

```python
import numpy as np
```

```python
x = np.arange(-3, 3, 0.1)
y = np.sin(x)
z = np.arange(3, -3, 0.1)
plt.plot(x, y, c='lightskyblue', label='y = sin(x)')
plt.plot([-3, 3], [-1, 1], c='lightcoral', label='y = 1/3x')
plt.plot([3, -3], [-1, 1], c='lightcoral', label='y = 1/3x')
plt.legend()
plt.title('Line Plot')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
area = np.pi * (15 * np.random.rand(N))**2
plt.title('Scatter Plot')
plt.scatter(x, y, s=area, alpha=0.5, c='gold')
plt.show()
```

```python
plt.hist(np.random.randn(1000), color='yellowgreen')
plt.title('Histgram')
plt.show()
```

```python
x = np.array([0.2, 0.6, 0.15, 0.25])
labels = ['Melon', 'Banana', 'Grape', 'Apple']
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
plt.pie(x, autopct='%d%%', labels=labels, colors=colors)
plt.axis('equal')
plt.title('Pie Chart')
plt.show()
```
