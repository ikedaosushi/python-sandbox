```python
import os
from pygeocoder import Geocoder
import googlemaps
from dotenv import load_dotenv
load_dotenv(".env");
```

```python
api_token = os.environ['GM_TOKEN']
client = googlemaps.Client(api_token)
```

```python
geocode_result = client.geocode(
    'Kaizen Platform', 
    region="jp", 
    language='ja'
)
loc = geocode_result[0]['geometry']['location']
```

```python
place_result = client.places_nearby(location=loc, radius=200, type='food', language='ja')
```
